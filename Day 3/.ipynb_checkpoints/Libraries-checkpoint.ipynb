Bar Chart Exercise

```python
categories = ["Company A", "Company B", "Company C", "Company D"]
values = [20,50,10,40]

colors = ["green", "blue", "red", "orange"] #choose colors to match each category

plt.bar(categories, values, color = colors)
plt.title("Bar chart")
plt.xlabel("Northwest companies")
plt.ylabel("Revenues")
plt.xticks(rotation = 45) #how much the title should be rotated

plt.show()    
```

# Create a 3d chart

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

categories = ["Company A", "Company B", "Company C", "Company D"]
values = [20,50,-11,40]

colors = ["green", "blue", "red", "orange"] #choose colors to match each category

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')   
"""
It's 1,1,1 that means a grid 1 by 1 and I specify that I'm plotting the element n.1
"""
x = np.arange(len(categories))
ax.bar(x, values, zdir='y', color=colors)   #I expect 4 elements


#plt.title("Bar chart")
#plt.xlabel("Companies")
#plt.ylabel("Values")
#plt.xticks(rotation = 45) #how much the title should be rotated

plt.show()   
```

```python
pivot = df1.pivot_table(
    index = "rooms",        # how many houses I have that have each number of rooms
    values = "city",        #I want to count how many cities do I have for 1 room, for 2 rooms, etc.
    aggfunc = "count") 
```

```python
pivot.plot(kind= "bar")
```

```python
max_price = df1["total (R$)"].max()     #maximum price and assign it to a variable
print(f"The maximum price of a house is: {max_price}")
```

```python
# top 10% of houses by price

threshold= max_price*0.9
print(f"The threshold is {threshold}")
```

```python
# I want a dataframe that contains only the uppest 10 % of price

top_df = df1[df1["total (R$)"]>threshold]
top_df
"""
There are outliers and we should remove them
"""
```

```python
df2 = df1[df1["total (R$)"]<900000]
df2
```

```python
df2 = df1[df1["total (R$)"]<1e4]
max_price2 = df2["total (R$)"].max()     #maximum price and assign it to a variable
print(f"The maximum price of a house is: {max_price}")
threshold= max_price2*0.9
print(f"The threshold is {threshold}")
```

```python
top_df2 = df2[df2["total (R$)"]>threshold]
top_df2
```

```python
top_df2_sorted = top_df2.sort_values(
    by = "total (R$)",
    ascending = False                 # it means I want it in descending order
)

top_df2_sorted.head(20)
```
