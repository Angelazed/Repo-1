```python
number = int(input("Please input number: "))
if number > 0:
    print("Positive")
elif number == 0:
    print("Zero")
else:
    print("Negative")
```

```python
age = int(input("Please input you age: "))
if age < 13:
    print("Child")
elif 13 <= age <= 19:
    print("Teenager")
elif 20 <= age <= 65:
    print("Adult")
else: 
    print("Senior")
```

```python
def introduce(name, age=25):
    print(f"Hello, this person is {name}, and they are {age} years old")

name = input("Please input your name: ")
age = input("Please input your age: ")

if age == "":
    introduce(name)
else:
    introduce(name, int(age))
```
