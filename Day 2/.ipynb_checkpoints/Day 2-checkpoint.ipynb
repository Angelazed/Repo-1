```python
def is_even(a):
    """
    This function returns True if number is even, 
    otherwise it returns False
    """
    result = a % 2
    if result == 1:
        return False
    else:
        return True

is_even(10)
```

```python
def is_even_2(a):
    """
    This function returns True if number is even, 
    otherwise it returns False
    """
    if a % 2 == 1:
        return False
    else:
        return True

is_even(10)
```

```python
is_even(11)
```

```python
is_even_2(11)
```

```python
def is_even_3(a):
    """
    This function returns True if number is even, 
    otherwise it returns False
    """
    if a % 2 == 0:
        return True

    return False

is_even_3(11)
```

```python
def is_even_4(a):
    return (a % 2 == 0) #returns True if the result is 0, otherwise returns False

is_even_4(13)
```
