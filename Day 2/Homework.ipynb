```python
value1 = input("Enter first boolean value: ").lower() == "true"
value2 = input("Enter second boolean value: ").lower() == "true"
log_op = str(input("Enter logical operator among and/or/not: "))

if log_op == "and":
    print(value1 and value2)
elif log_op == "or":
    print(value1 or value2)
elif log_op == "not":
    print(not value1)
else:
    print("Invalid inputs")
```
